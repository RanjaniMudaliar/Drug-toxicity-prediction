# Toxicity

```python
#data manipulation
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("descriptors 1.csv")
```

```python
data
```

```python
data = data.fillna(0)
```

```python
data
```

```python
cols = data.shape[1]
x = data.iloc[:,0:cols-1] #separating the input from data set
y = data.iloc[:,cols-1:cols] #separating the output from data set
```

```python
x
```

```python
y
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.1, random_state = 42) #depends on test_size
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 100, random_state = 1000) #depends on randomness

# Train the model on training data
rf.fit(xtrain, ytrain);
```

```python
Predictions = rf.predict(xtest)
```

```python
Predictions
```

```python
#Regression score function
from sklearn.metrics import r2_score
r2_score(ytest , Predictions)
```

# Ensemble Learning Model

```python
from mlxtend.regressor import StackingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
#construct the models
lr = LinearRegression()
svr_lin = SVR(kernel='linear')
clf1 = KNeighborsRegressor(n_neighbors=11)
clf2 = RandomForestRegressor(n_estimators = 100, random_state=20000, max_features='auto',min_samples_leaf=3)
svr_rbf = SVR(kernel='rbf')

stregr = StackingRegressor(regressors=[svr_lin, lr, clf1, clf2], 
                           meta_regressor=svr_rbf)
```

```python
#model fit
reg = stregr.fit(xtrain, ytrain)
pred = reg.predict(xtest)
```

```python
pred
```

```python
#Regression score function
from sklearn.metrics import r2_score
r2_score(ytest , pred)
```
